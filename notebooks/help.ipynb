```python
from transformers import AutoModel, AutoConfig
# Save the CONFIG only (since you overwrite weights later anyway)
config = AutoConfig.from_pretrained('bert-base-uncased')
config.save_pretrained('./src/bert_config')

# OR save the full model if you want strict loading
model = AutoModel.from_pretrained('bert-base-uncased')
model.save_pretrained('./src/bert_base')
```

```python
import torch
from torchvision import models
import os

# Create a folder for it
save_path = "models/resnet50"
os.makedirs(save_path, exist_ok=True)

# 1. Download the model with weights
print("Downloading ResNet50...")
model = models.resnet50(pretrained=True)

# 2. Save ONLY the weights (state_dict)
torch.save(model.state_dict(), os.path.join(save_path, "resnet50.pth"))
print(f"Saved to {save_path}/resnet50.pth")
```
